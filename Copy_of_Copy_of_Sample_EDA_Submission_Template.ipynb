# <b> Step2: Handling missing values. </b>

````python
# Columns having missing values.
# To address the NaN values in the dataset, we first need to identify the columns that contain null values.
new_df.isnull().sum().sort_values(ascending = False) [:5]
````

Since, company and agent columns have so many number and agent numbers as data. There may be some cases when customer didnt booked hotel via any agent or via any company. So in that case values can be null under these columns.
We will replace null values by 0 in these columns

````python
# As we have seen, there is minimul null values in agent, Lets fill these value by 0
````

````python
new_df[['company','agent']] = new_df[['company','agent']].fillna(0)
````

````python
new_df['children'].unique()
````

This column 'children' has 0 as value which means 0 children were present in group of customers who made that transaction.
So, 'nan' values are the missing values due to error of recording data.


We will replace the null values under this column with mean value of children.

````python
# As we have seen, there is minimul null values in children, Lets fill these value by taking mean of the all values
````

````python
new_df['children'].fillna(new_df['children'].mean(), inplace = True)
````


Next column with missing value is 'country'. This column represents the country of origin of customer.
Since, this column has datatype of string. We will replace the missing value with the mode of 'country' column.

````python
new_df['country'].fillna('others', inplace = True)
````

````python
# Checking if all null values are removed
new_df.isnull().sum().sort_values(ascending = False)
````

There are some rows with total number of adults, children or babies equal to zero. So we will remove such rows.

````python
new_df[new_df['adults']+new_df['babies']+new_df['children'] == 0].shape
````

````python
new_df.drop(new_df[new_df['adults']+new_df['babies']+new_df['children'] == 0].index, inplace = True)
````

# Step 3: Converting columns to appropriate datatypes.

````python
# Converting datatype of columns 'children', 'company' and 'agent' from float to int.
# We have seen that children & agent column as datatype as float whereas it contains only int value, lets change datatype as 'int64'
new_df[['children', 'company', 'agent']] = new_df[['children', 'company', 'agent']].astype('int64')
````

````python
# changing datatype of column 'reservation_status_date' to data_type.
new_df['reservation_status_date'] = pd.to_datetime(new_df['reservation_status_date'], format = '%Y-%m-%d')
````

# Step 4: Adding important columns.

````python
# Adding total staying days in hotels
new_df['total_stay'] = new_df['stays_in_weekend_nights']+new_df['stays_in_week_nights']
````

````python

# Adding total people num as column, i.e. total people num = num of adults + children + babies
new_df['total_people'] = new_df['adults']+new_df['children']+new_df['babies']
````

Next, we will investigate whether not getting allotted the same room type as demanded is a potential cause of booking cancellations.
